```python
import pandas as pd
```

```python
def read_owu(file, root_path = 'dataset/datahow_2020/insilico_data'):
	data = pd.read_csv(f'{root_path}/{file}.csv')
	owu_df = data.copy()
	owu_df.set_index(['run', 'time'], inplace=True)
	return owu_df
```

```python
root_path = 'dataset/datahow_2020/insilico_data/interpolation'
owu = read_owu('owu', root_path=root_path)
owu_test = read_owu('owu_test', root_path=root_path)
```
